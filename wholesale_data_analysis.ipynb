## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df['subtotal'] = df['unit_price'] * df['qty']
df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df['total_weight'] = df['unit_weight'] * df['qty']
df['shipping_price'] = np.where(
    df['total_weight'] > 50,
    df['total_weight'] * 7,  # if weight > 50
    df['total_weight'] * 10  # if weight ≤ 50
)
print("Verifying shipping price calculations:")
sample_rows = df.head()
print(pd.DataFrame({
    'Weight': sample_rows['total_weight'],
    'Shipping Price': sample_rows['shipping_price']
}))
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
def calculate_total_price(df):
    df['total_price'] = (df['subtotal'] + df['shipping_price']) * .0925
    return df
print(df.columns)
print(df.head)
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
def calculate_line_cost(row):
    base_cost = row['unit_cost'] * row['qty']
    shipping = row['unit_weight'] * row['qty'] * 0.50  # Example shipping rate
    return base_cost + shipping

df['total_line_cost'] = df.apply(calculate_line_cost, axis=1)
print(df.columns)
print(df.head)

```

```python
# Create a column for the profit of each line using line cost and line price
def calculate_profit(row):
    line_revenue = row['unit_price'] * row['qty']
    return round(line_revenue - row['total_line_cost'], 2)

df['line_profit'] = df.apply(calculate_profit, axis=1)
print(df.columns)
print(df.head)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.

# Get all orders for the top five clients (in client orders)
def calculate_client_spending(client_id):
    client_orders = df[df['client_id'] == client_id]
    
    # Calculate each of the totals (sum of Price+orders, shipping price, shipping + subtotal, calc &+ tax =total)
    subtotal = (client_orders['unit_price'] * client_orders['qty']).sum()
    shipping_total = client_orders['shipping_price'].sum()
    pre_tax_total = subtotal + shipping_total
    total_with_tax = pre_tax_total * .0925 
    
    # Get client name (create client name var. & assign 1st & last manes using a locate fx)
    client_name = f"{client_orders['first'].iloc[0]} {client_orders['last'].iloc[0]}"
    
    return {
        'client_id': client_id,
        'name': client_name,
        'total_spent': total_with_tax
    }

# Calculate spending for each top client
top_client_spending = []
for client_id in top_client_ids:
    spending = calculate_client_spending(client_id)
    top_client_spending.append(spending) 

top_client_spending = sorted(top_client_spending, key=lambda x: x['total_spent'], reverse=True)


# Create a formatted display of results
print("Our Top Five Clients (by Quantity of products) Total Spending has been:")
print("_" * 27)
for client in top_client_spending:
    print(f"Client ID: {client['client_id']}")
    print(f"Name: {client['name']}")
    print(f"Total Spent: ${client['total_spent']:,.2f}")
    print("_" * 27)
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
summary_data = []
for client_id in top_client_ids:
    # ID all orders for each client
    client_orders = df[df['client_id'] == client_id]
   
    # then calculate totals
    client_summary = {
        'Client': client_id,
        'Name': f"{client_orders['first'].iloc[0]} {client_orders['last'].iloc[0]}",
        'Units Purchased': client_orders['qty'].sum(),
        'total_shipping': client_orders['shipping_price'].sum(),
        'total_revenue': (client_orders['unit_price'] * client_orders['qty']).sum(),
        'total_profit': client_orders['line_profit'].sum()
    }
    summary_data.append(client_summary)

# create DF from summary data
summary_df = pd.DataFrame(summary_data)

# Format currency columns
currency_cols = ['total_shipping', 'total_revenue', 'total_profit']
for col in currency_cols:
    summary_df[col] = summary_df[col].apply(lambda x: f"${x:,.2f}")

# Display Summary output with improved formatting
print("Top 5 Clients Summary:")
print(summary_df.to_string(
    index=False,
    justify='left',     
    formatters={        
        'Units Purchased': lambda x: f'{x:>10}',
        'total_shipping': lambda x: f'{x:>12}',
        'total_revenue': lambda x: f'{x:>12}',
        'total_profit': lambda x: f'{x:>12}'
    }
))
```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
money_columns = {
    'total_shipping': 'Total Shipping',
    'total_revenue': 'Total Revenue', 
    'total_profit': 'Total Profit'
}

def currency_format_millions(value_str):
    value = float(value_str.replace('$', '').replace(',', ''))
    value_millions = value / 1000000
    return f"${value_millions:.2f}M"

# Create a copy of the summary DataFrame
formatted_df = summary_df.copy()

# Apply currency formatting
for col in money_columns.keys():
    formatted_df[col] = formatted_df[col].apply(currency_format_millions)

# Define the new column names (fixing the spelling)
column_rename = {
    'Client ID#': 'Client',
    'Client Name': 'Name',
    'Total Units Purchased': 'Units Purchased',  # Fixed spelling
    'total_shipping': 'Total Shipping (millions)',
    'total_revenue': 'Total Revenue (millions)',
    'total_profit': 'Total Profit (millions)'
}

# Rename the columns
formatted_df = formatted_df.rename(columns=column_rename)

# Sort by profit
sorted_df = formatted_df.sort_values(
    by='Total Profit (millions)',
    ascending=False,
    key=lambda x: x.str.replace('$', '').str.replace('M', '').astype(float)
)

# Configure pandas display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Print formatted output
print("Clinet Spending Summary Report (Sorted by Total Profit)")

# Create col_space dictionary using the correct column names
col_space = {
    'Client': 10,
    'Name': 20,
    'Units Purchased': 15,  # Fixed spelling
    'Total Shipping (millions)': 27,
    'Total Revenue (millions)': 27,
    'Total Profit (millions)': 27
}

# Print the DataFrame with custom formatting
print(sorted_df.to_string(
    index=False,
    justify='left',
    col_space=col_space
))
```
